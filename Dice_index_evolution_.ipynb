```python
import matplotlib.pyplot as plt

# Data points
number_of_images = [2, 4, 8, 16, 32, 65]
dice_index = [0.8292162873598504, 0.7496580081093523, 0.8878475694122515, 0.9074998415870354, 0.9064258692946444, 0.9174153059005076]

# Create the plot
plt.plot(number_of_images , dice_index , marker='*', linestyle='-')

# Add labels and title
plt.xlabel("Number of images")
plt.ylabel("Dice index ")
plt.title("Evolution of the Dice index according to  the number of images")

# Show the plot
plt.grid(True)
plt.show()
```
